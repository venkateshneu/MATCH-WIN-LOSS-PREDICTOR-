```python
total_runs = df.groupby(["match_id","inning"])["total_runs"].sum().reset_index()
```

```python


# Grouping the DataFrame by "match_id" and "inning", and calculating the sum of "total_runs"
total_runs = df.groupby(["match_id", "inning"])["total_runs"].sum().reset_index()


```

```python
total_runs.head(2)
```

```python
# Filtering the DataFrame total_runs to include only rows where the value in the "inning" column is equal to 1
total_runs = total_runs[total_runs["inning"] == 1]
```

```python
total_runs
```

```python
# Merging the DataFrames dp and total_runs based on the "id" column of dp and the "match_id" column of total_runs,
# and including only the "match_id" and "total_runs" columns from total_runs in the resulting DataFrame
matches = dp.merge(total_runs[["match_id", "total_runs"]], left_on="id", right_on="match_id")
```

```python
matches.head(2)
```

```python
matches["team1"].unique()
```

```python
matches.shape
```

```python
teams = ['Sunrisers Hyderabad', 'Mumbai Indians', 
        'Royal Challengers Bangalore',
       'Kolkata Knight Riders',  'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals','Delhi Capitals']
```

```python
# Replacing 'Delhi Daredevils' with 'Delhi Capitals' in the "team1" column
matches["team1"] = matches["team1"].str.replace('Delhi Daredevils', 'Delhi Capitals')

# Replacing 'Delhi Daredevils' with 'Delhi Capitals' in the "team2" column
matches["team2"] = matches["team2"].str.replace('Delhi Daredevils', 'Delhi Capitals')

# Replacing 'Deccan Chargers' with 'Sunrisers Hyderabad' in the "team1" column
matches["team1"] = matches["team1"].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')

# Replacing 'Deccan Chargers' with 'Sunrisers Hyderabad' in the "team2" column
matches["team2"] = matches["team2"].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
# Filtering the DataFrame matches to include only rows where "team1" is in the list of specified teams
matches = matches[matches["team1"].isin(teams)]

# Further filtering the DataFrame matches to include only rows where "team2" is also in the list of specified teams
matches = matches[matches["team2"].isin(teams)]
```

```python
matches["team2"].unique()
```

```python
matches.shape
```

```python
# Filtering the DataFrame matches to include only rows where the value in the "dl_applied" column is equal to 0
matches = matches[matches["dl_applied"] == 0]
```

```python
matches.shape
```

```python
# Selecting specific columns from the DataFrame matches
matches = matches[["match_id", "city", "winner", "total_runs"]]
```

```python
delivery = matches.merge(df, on ="match_id" )
```

```python
delivery.head(2)
```

```python
delivery = delivery[delivery["inning"]==2]
```

```python
delivery.shape
```

```python
# Renaming the column "total_runs_x" to "final_score" in the DataFrame delivery
delivery.rename(columns={"total_runs_x": "final_score"}, inplace=True)
```

```python
# Calculating the cumulative sum of the "total_runs_y" column within each group defined by the "match_id" column
# and assigning the result to a new column named "present_score"
delivery["present_score"] = delivery.groupby("match_id")["total_runs_y"].cumsum()
```

```python
delivery.head()
```

```python
# Calculating the remaining score by subtracting the present score from the final score
delivery["remaining_score"] = delivery["final_score"] - delivery["present_score"]
```

```python
# Calculating the number of balls remaining in the innings
delivery["balls_remaining"] = 126 - (delivery["over"] * 6 + delivery["ball"])
```

```python
delivery.head()
```

```python
# Filling missing values in the "player_dismissed" column with "0"
delivery["player_dismissed"] = delivery["player_dismissed"].fillna("0")

# Mapping non-missing values in the "player_dismissed" column to "1" and keeping missing values as "0"
delivery["player_dismissed"] = delivery["player_dismissed"].apply(lambda x: x if x == "0" else "1")

# Converting the "player_dismissed" column to integer type
delivery["player_dismissed"] = delivery["player_dismissed"].astype(int)

# Calculating the cumulative sum of player dismissals within each match
wickets = delivery.groupby("match_id")["player_dismissed"].cumsum().values

# Calculating the number of wickets remaining in the innings
delivery["wickets_rem"] = 10 - wickets
```

```python
# current runrate = totalruns/overs
```

```python
delivery["crr"] = (delivery["present_score"]*6)/( 120 - delivery["balls_remaining"])
```

```python
delivery["rrr"] = (delivery["remaining_score"]*6)/ delivery["balls_remaining"]
```

```python
delivery.shape
```

```python
def result(row):
    return 1 if row["winner"] == row["batting_team"] else 0
```

```python
delivery["result"] = delivery.apply(result, axis = 1)
```

```python
delivery.shape
```

```python
delivery = delivery[["batting_team", "bowling_team", "city","remaining_score", "balls_remaining", "wickets_rem", "final_score", "crr", "rrr", "result",   ]]
```

```python
delivery = delivery.sample(delivery.shape[0])
```

```python
delivery
```

```python
# Removing rows with missing values from the DataFrame delivery
delivery.dropna(inplace=True)
```

```python
# Filtering the DataFrame delivery to include only rows where the value in the "balls_remaining" column is not equal to 0
delivery = delivery[delivery["balls_remaining"] != 0]
```

```python
from sklearn.model_selection import train_test_split

# Assigning features (X) and target variable (y)
X = delivery.iloc[:, :-1]  # All columns except the last one
y = delivery.iloc[:, -1]   # Last column

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
delivery.isnull().sum()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python


# Defining a ColumnTransformer named trf
trf = ColumnTransformer([
    # Applying OneHotEncoder to the specified columns with options sparse_output=False and drop="first"
    ("trf", OneHotEncoder(sparse_output=False, drop="first"), ["batting_team", "bowling_team", "city"])
], 
# Passthrough all remaining columns without any transformation
remainder="passthrough")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
# Defining a Pipeline named pipe
pipe = Pipeline([
    # The first step in the pipeline applies the ColumnTransformer trf
    ("trf1", trf),
    # The second step in the pipeline applies a LogisticRegression model with the specified solver
    ("trf2", LogisticRegression(solver="liblinear"))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
X_test
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Define the pipeline with GradientBoostingClassifier
pipe_gb = Pipeline([
    ("trf1", trf),  # First step: Apply ColumnTransformer
    ("gb", GradientBoostingClassifier())  # Second step: Apply GradientBoostingClassifier
])

# Fit the pipeline to the training data
pipe_gb.fit(X_train, y_train)

# Make predictions on the test data
predictions_gb = pipe_gb.predict(X_test)

# Calculate accuracy
accuracy_gb = accuracy_score(y_test, predictions_gb)
print("Accuracy with GradientBoostingClassifier:", accuracy_gb)
```
